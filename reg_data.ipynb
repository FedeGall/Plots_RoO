```python
import time
import pandas as pd
import numpy as np
import re
from utilities import *
```

```python
path_temp = Path('temp/')
path_data = Path('data/')
```

```python
baci = pd.read_csv(path_temp / "kor_full.csv", dtype = {"product_code": "str", "origin": "str", "destination": "str"})
```

```python
baci['product_code'] = baci['product_code'].apply(lambda x: '0' + x if len(x) == 5 and re.match(r'[0-9]{5}', x) else x)
baci = baci[baci["product_code"].str.contains(r'[0-9]{6}', regex=True, na=False)]
baci = baci[~(baci["origin_name"].isin(["EU27", "EU28"]))]
baci = baci[~(baci["destination_name"].isin(["EU27", "EU28"]))]
baci.to_csv(path_data / "covfefe.csv", index = False)
```
