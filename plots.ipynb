```python
import dash
import dash_core_components as dcc
import dash_html_components as html
import dash_bootstrap_components as dbc
from dash.dependencies import Input, Output
import plotly.express as px
import pandas as pd
from pathlib import Path

input_path = Path('temp/kor_full.csv')
df = pd.read_csv(input_path)
importers = set(df['destination_name'].unique())
exporters = set(df['origin_name'].unique())
countries = list(importers.union(exporters))
```

```python
app = dash.Dash(__name__)

app.layout = html.Div([
    dcc.Dropdown(
        id="trade_partner",
        options=[{"label": x, "value": x} 
                 for x in countries],
        value='France',
    ),
    dcc.Graph(id="line-chart"),
])

@app.callback(
    Output("line-chart", "figure"), 
    [Input("trade_partner", "value")])
def update_line_chart(country):
    filtered_df = df[(df['origin_name'] == country) | (df['destination_name'] == country)]
    fig = px.line(filtered_df, x="year", y="value", color='trade_flow')
    return fig

app.run_server(debug=True)
```
